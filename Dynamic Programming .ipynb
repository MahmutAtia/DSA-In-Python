```python
#Try 3 optimized
def travle3(w,h, memo = {}):
  if w == 1 and h == 1:
    return 1
  if w == 0 or h == 0:
    return 0

  if not memo.get((w-1,h)):
    memo[(w-1,h)] = travle3(w-1,h)
  if not memo.get((w,h-1)):
    memo[(w,h-1)] = travle3(w,h-1)


  return memo[(w-1,h)] + memo[(w,h-1)]
```

5- canSum

```python
def canSum(n, li):
  print(n)
  if n==0:
    return True
  if n < 0:
    return False


  for i in li:
     if canSum(n-i,li):
        return True
  return False

```
