```python
def fin(w, h, area, total=0, count=0, li=None, min_count=float('inf'), min_config=None):
    # Initialize the list if it's None
    if li is None:
        li = []

    # Base cases
    if total > area:
        return False, min_count, min_config
    if area == total:
        if count < min_count:
            min_count = count
            min_config = li.copy()
        return True, min_count, min_config

    # Explore all possible configurations recursively
    for num in range(min(w, h), 0, -1):
        new_total = total + num ** 2
        if new_total <= area:
            li.append(num)
            success, min_count, min_config = fin(num, h, area, new_total, count + 1, li, min_count, min_config)
            li.pop()
            if success and min_count < float('inf'):
                return True, min_count, min_config

    return False, min_count, min_config

# Example usage
width = 7
height = 20
area_to_fill = 140

success, min_count, min_config = fin(width, height, area_to_fill)
if success:
    print(f"Minimum count of squares needed: {min_count}")
    print(f"Configuration of squares: {min_config}")
else:
    print("No valid configuration found.")
```

```python
def is_valid(x, y, size, filled):
    # Check if the square at (x, y) with given size is within bounds and not filled
    for i in range(x, x + size):
        for j in range(y, y + size):
            if (i, j) in filled:
                return False
    return True

def fill_squares(width, height, filled, count):
    if len(filled) == width * height:  # All squares filled
        return count

    min_count = float('inf')
    for i in range(width):
        for j in range(height):
            for size in range(1, max(width, height) + 1):
                if is_valid(i, j, size, filled):
                    for x in range(i, i + size):
                        for y in range(j, j + size):
                            filled.add((x, y))  # Mark squares as filled
                    min_count = min(min_count, fill_squares(width, height, filled, count + 1))
                    for x in range(i, i + size):
                        for y in range(j, j + size):
                            filled.remove((x, y))  # Backtrack: Unmark squares

    return min_count

def min_squares_to_fill(width, height):
    return fill_squares(width, height, set(), 0)

# Example usage
width = 10
height = 7
min_squares = min_squares_to_fill(width, height)
print(f"Minimum number of squares to fill {width}x{height} area: {min_squares}")
```

```python
def min_squares_to_fill(width, height):
    # Base case: If either width or height is 0, return 0 squares needed
    if width == 0 or height == 0:
        return 0

    # If the width and height are equal, it's a square, so only 1 square is needed
    if width == height:
        return 1

    min_count = float('inf')

    # Try all possible vertical cuts and horizontal cuts to find the minimum
    for i in range(1, width):
        # Vertical cut
        count = 1 + min_squares_to_fill(i, height) + min_squares_to_fill(width - i, height)
        min_count = min(min_count, count)

    for j in range(1, height):
        # Horizontal cut
        count = 1 + min_squares_to_fill(width, j) + min_squares_to_fill(width, height - j)
        min_count = min(min_count, count)

    return min_count

# Example usage
width = 10
height = 7
min_squares = min_squares_to_fill(width, height)
print(f"Minimum number of squares to fill {width}x{height} area: {min_squares}")
```

```python
# Python3 program to find minimum
# number of squares
# to cut a paper using Dynamic Programming

MAX = 300
dp = [[0 for i in range(MAX)] for i in range(MAX)]

# Returns min number of squares needed


def minimumSquare(m, n):

	# Initializing max values to
	# vertical_min
	# and horizontal_min
	vertical_min = 10000000000
	horizontal_min = 10000000000

	# N=11 & M=13 is a special case
	if n == 13 and m == 11:
		return 6
	if m == 13 and n == 11:
		return 6

	# If the given rectangle is
	# already a square
	if m == n:
		return 1

	# If the answer for the given rectangle is
	# previously calculated return that answer
	if dp[m][n] != 0:
		return dp[m][n]

	# The rectangle is cut horizontally and
	# vertically into two parts and the cut
	# with minimum value is found for every
	# recursive call.
	for i in range(1, m//2+1):

		# Calculating the minimum answer for the
		# rectangles with width equal to n and length
		# less than m for finding the cut point for
		# the minimum answer
		horizontal_min = min(minimumSquare(i, n) +
							minimumSquare(m-i, n), horizontal_min)
	for j in range(1, n//2+1):

		# Calculating the minimum answer for the
		# rectangles with width equal to n and length
		# less than m for finding the cut point for
		# the minimum answer
		vertical_min = min(minimumSquare(m, j) +
						minimumSquare(m, n-j), vertical_min)

	# Minimum of the vertical cut or horizontal
	# cut to form a square is the answer
	dp[m][n] = min(vertical_min, horizontal_min)
	return dp[m][n]


# Driver code
if __name__ == '__main__':
	m = 7
	n = 20
	# Function call
	print(minimumSquare(m, n))

# This code is contributed by sahilshelangia
```
