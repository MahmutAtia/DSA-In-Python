```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val, prev, next, child):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child
"""

class Solution:
    def flatten(self, head: 'Optional[Node]') -> 'Optional[Node]':
        curr = head
        last_heads = []

        while curr:

            #check
            print("adding the last head ", curr.val, "previous is " ,  curr.prev.val if curr.prev else "No Prev", "next is " , curr.next.val if curr.next else "NO Next")

            # is there a child?
            if curr.child:
                last_heads.append(curr.next) # add the head
                print("the new head is" , last_heads[-1].val)
                curr.next = curr.child
                curr.next.prev = curr # add prev to child

                curr = curr.next # move the loop on


            else:
                if curr.next == None:
                    if last_heads:
                        curr.next = last_heads.pop()
                        curr.next.prev = curr #add prev to the new head
                        curr = curr.next


                    else:
                        curr =curr.next
                else:
                    curr = curr.next

        return head
```
