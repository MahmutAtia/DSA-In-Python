```python
class MyQueue:

    def __init__(self):
        self.stack= []
        self.temp = []


    def push(self, x: int) -> None:
        self.stack.append(x)


    def pop(self) -> int:
        if self.temp:
            return self.temp.pop()
        else:
            while len(self.stack)> 1:
                ele = self.stack.pop()
                self.temp.append(ele) # is a queue now


            return self.stack.pop() #last element

    def peek(self) -> int:
        if self.temp:
            return self.temp[-1]
        else:
            return self.stack[0]


    def empty(self) -> bool:
        if not self.stack and not self.temp:
            return True
        else:
            return False



# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
