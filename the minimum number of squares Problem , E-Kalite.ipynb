### Problem Definition
#### Given a rectangle of dimensions width and height, *the* task is to determine the minimum number of squares required to completely fill the rectangle without overlapping or leaving any gaps.

Birinci Deneme:
Recursion kullandım , dün çıkan sonuç elde ettim, zaten 9dan aşağı sonuç alamadık

```python
class Solution:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def __call__(self):
        wh_li = [self.w, self.h]
        count = self._recursive(wh_li)
        return count

    def _define_wh(self, li):
        if li[0] > li[1]:
            return li[1], li[0]
        elif li[0] < li[1]:
            return li[0], li[1]
        else:
            return li[0], li[1]

    def _recursive(self, wh_li, li=[], count=0, min_count=float('inf')):
        w, h = self._define_wh(wh_li)

        # Base case, when to stop
        if w * h == 0:
            print('Square List is : ',li)  # printing solution
            return count

        li.append(w)
        new_wh_li = (w, h - w)
        return self._recursive(new_wh_li, li, count + 1, min_count)



```

```python
# Creating an instance of Solution
solution_instance = Solution(7, 20)

# Calling the __call__ method
result = solution_instance()
print("Square Count is : ", result)
```

İkinci Deneme:
Area baz alarak backtracking ve recursion yaptım. Sonuç matematik olarak doğru da ama çizince olmuyor

```python
def backtrack(w, h, area, total=0, count=0, li=None):
    if not li:  # because the function will append to the same list if called multiable time , it the same in the memory
      li = []

    # Base cases
    if total > area:
        return False
    if area == total:

        print("Squares are : ", li)
        print("Square Count is : ", count)
        return True

    # which is smaller
    w = min(w,h)
    h = max(w,h)

    for num in range(w, 0, -1):
        li.append(num)
        if backtrack(num, h, area, total + num ** 2, count + 1, li):
            return True
        else:
            li.pop()

    return False
```

```python

width = 20
height = 7
total_area = width * height

result = backtrack(width, height, total_area)
```

```python
class Solution:
    def tilingRectangle(self, n: int, m: int) -> int:
		board = [[False for _ in range(m)] for _ in range(n)]
            self.min_n_tiles = float('inf')

            def find_next_empty_cell(board):
				# go over the entire board, find the first empty cell and return it
                for i in range(n):
                    for j in range(m):
                        if not board[i][j]:
                            return i, j

				# if the board is all covered, reutrn -1,-1
                return -1, -1

            def get_max_tile_size(board, row, col):
                # we need to measure the min distance from row, col to a tile/wall
                row_dist = col_dist = 1

                while col + col_dist < m and not board[row][col + col_dist]: col_dist += 1
                while row + row_dist < n and not board[row + row_dist][col]: row_dist += 1

                return min(row_dist, col_dist)

            def cover(board, row, col, tile_size, cover_val):
				# cover board at specific location with a specific tile, by the cover val argument (True/False)
                for i in range(tile_size):
                    for j in range(tile_size):
                        board[row + i][col + j] = cover_val

            def backtrack(board, n_tiles):
				# if curr tile count is already the min_tiles we found so far, no need to continue this placing route
                if n_tiles == self.min_n_tiles:
                    return

                next_row, next_col = find_next_empty_cell(board)

                # if not empty cell -> update min and break
                if next_row == next_col == -1:
                    self.min_n_tiles = min(self.min_n_tiles, n_tiles)
                    return

                # find max tile size to place
                max_tile_size = get_max_tile_size(board, next_row, next_col)

                # iterate possible tile size, place and backtrack
                for tile_size in range(max_tile_size, 0, -1):
                    # place tile
                    cover(board, next_row, next_col, tile_size, True)

                    # backtrack
                    backtrack(board, n_tiles + 1)

                    # remove tile
                    cover(board, next_row, next_col, tile_size, False)


            backtrack(board, 0)
            return self.min_n_tiles
```
