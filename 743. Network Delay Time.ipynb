```python
class Solution:
    def networkDelayTime(self, times: List[List[int]], n: int, k: int) -> int:
        adj_li = self.adj_li(times,n)

        # Bell ford algorithm
        head = k-1
        costs = [float("inf") if i != head else 0 for i in range(n) ]




        for i in range(n-1): # bell ford iter
            for head, nodes in enumerate(adj_li):
                for li in nodes:
                    target = li[0]
                    w = li[1]

                    if (costs[head] + w) < costs[target]:
                        costs[target] = costs[head] + w





        if float("inf") in costs:
            return -1
        else:
            return max(costs)
    def adj_li(self,times,n):
        li = [[] for _ in range(n)]
        for time in times:
            source = time[0]-1
            target = time[1]-1
            w = time[2]
            li[source].append([target,w])

        return li
```
