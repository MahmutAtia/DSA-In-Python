```python
class Solution(object):
    def fib(self, n):
        """
        :type n: int
        :rtype: int
        """
        return  self.f(n)

    def f(self,n):
        if n == 1:
            return 1
        if n == 0:
            return 0

        # print(n-1,n-2)
        out = self.f(n-1) + self.f(n-2)
        # print("out", out)
        return out
```
