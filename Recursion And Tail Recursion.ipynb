Factorial Func with recursion O(n)




```python
def fact(x):
  if x <= 1:
    return 1


  return (x * fact(x-1))
```

```python
fact(3)
```

Factorial Func with tail recursion O(1)


```python
def fact_O1(x,totlal_so_far =1):
  if x==0:
    return totlal_so_far

  totlal_so_far = totlal_so_far * x
  return fact_O1(x-1,totlal_so_far)
```

Try 2:

```python
def fact_O1_2(x,totlal_so_far=1):
  if x == 0:
    return totlal_so_far
  else:
    return fact_O1_2(x-1,totlal_so_far*x)
```

```python
fact_O1_2(5)
```
