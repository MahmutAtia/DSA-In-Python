```python
def sort_for(array):
  length = len(array)
  for i in range(length):
    for j in range(length):

      if array[i] > array[j]:
        #swap indexes
        temp = array[i]
        array[i] = array[j]
        array[j] = temp

  return array
```

#### Selection Sort


```python
def select_sort(array):
  length = len(array)
  for i in range(length):
    minm = i
    for j in range(length):
      if array[j] < array[minm]:
        minm = j
    temp = array[i]
    array[i] = array[minm]
    array[minm] = temp

  return array
```

```python
def merge(left,right):
  left_idx = 0
  right_idx = 0
  result = []

  while left_idx < len(left) and right_idx < len(right) :
    if left[left_idx] > right[right_idx]:
      result.append(right[right_idx])
      left_idx +=1

    else:
      result.append(left[left_idx])
      right_idx +=1

    while left_idx < len(left):
        result.append(left[left_idx])
        left_idx +=1
    while right_idx < len(right):
      result.append(right[right_idx])
      right_idx +=1
    print(result)
  return result


def merge_sort(arr):
  if len(arr) == 1:
    return arr

  idx = len(arr) // 2
  left = arr[:idx]
  right = arr[idx:]
 # print(left,right)




  return merge(merge_sort(left), merge_sort(right))


```
