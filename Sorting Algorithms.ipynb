```python
def sort_for(array):
  length = len(array)
  for i in range(length):
    for j in range(length):

      if array[i] > array[j]:
        #swap indexes
        temp = array[i]
        array[i] = array[j]
        array[j] = temp

  return array
```

#### Selection Sort


```python
def select_sort(array):
  length = len(array)
  for i in range(length):
    minm = i
    for j in range(length):
      if array[j] < array[minm]:
        minm = j
    temp = array[i]
    array[i] = array[minm]
    array[minm] = temp

  return array
```
