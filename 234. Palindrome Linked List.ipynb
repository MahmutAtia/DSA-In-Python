```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:

        if not head.next:
            return True
        return self.f(head)
    def f(self,head,li=[]):
        if head:
            print(head.val)

        if not head:
            print(li,len(li))
            if not li:
                return True
            else:
                return False


        if li and li[-1] == head.val:
            li.pop()
        else:
             li.append(head.val)
        return self.f(head.next,li)

```
