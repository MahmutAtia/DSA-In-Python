```python
class Solution(object):
    def findArray(self, pref):
        db = [None] * len(pref)
        self.backtrack(pref, db, 0)
        return db if db[-1] is not None else []

    def backtrack(self, pref, db, n):
        if n == len(pref):
            return True

        for i in range(10**7):  # Try all possible values from 0 to 255
            db[n] = i

            if self.get_xor_pref(db, n) == pref[n]:

                if self.backtrack(pref, db, n + 1):
                    return True

        db[n] = None
        return False

    def get_xor_pref(self, db, n):
        total = 0
        for i in range(n + 1):
            if db[i] is None:
                break
            total ^= db[i]
        return total
```
