```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        all_possible_areas = []
        for indx1 , i1 in enumerate(height):
            round_areas = []
            for indx2,i2 in enumerate(height):
                w = indx2 - indx1
                h = min(i1,i2)
                round_areas.append(w*h)
            all_possible_areas.append(max(round_areas))
            round_areas = []

        return max(all_possible_areas)
```

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        area = 0
        for indx1 , i1 in enumerate(height):
            for indx2,i2 in enumerate(height):
                w = indx2 - indx1
                h = min(i1,i2)
                if w*h > area :
                    area = w*h

        return area
```

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        map = {}
        area = 0
        for p in range(len(height)):
            map[p] = height[p]

            if map.get(p) is not None:
                i2 = p
                while i2 >= 0:
                    h = min( height[p],map[i2] )
                    w =  p - i2

                    a = h*w
                    area = max(a,area)
                    i2 -=1

        return area
```
