```python
import pdb


class Array:
  def __init__(self):
    self.data = {}
    self.length = 0


  def push(self,item):
    self.data[self.length] = item
    self.length += 1

  def get(self,index):
    return self.data[index]

  def pop(self):
    self.length -= 1
    return self.data.pop(self.length) # after we substract 1

  def __repr__(self):
    return str( self.data.values())


  def delete(self,index):
    self.shift_indexes(index)
    self.data.pop(self.length-1)
    self.length = self.length -1






  def shift_indexes(self,index):
    for i in range(index,self.length -1):

      print(i,i+1, self.length)
      print(self.data[i+1])

      self.data[i] = self.data[i+1]
      print(self.data)




  def clear(self):
    self.data= {}
    self.length = 0









```
