```python
class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.

        """

        # loop over the grid
        for i in range(9):
            for j in range(9):
                if board[i][j] != ".":
                    continue

                else:
                    for num in range(1,10): # to fill it with a nuumber from 1to9
                        print(num)
                        num = str(num)
                        if self.search_in_seq(board,i,j,num):
                            print("pass seq")
                            continue

                        if num in board[i]:
                            print("pass i")

                            continue

                        if num in [board[row][j] for row in range(9) ]:
                            print("pass j")

                            continue

                        board[i][j] = num
                        print("--------------num found ---------")
                        break

        for b in board:
            print(b)




```
