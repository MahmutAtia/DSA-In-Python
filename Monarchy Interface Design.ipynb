```python
class T_node:
  def __init__(self,name):
    self.name = name
    self.is_alive = True
    self.children = []

  def new_birth(self, name):
    child = T_node(name)
    self.children.append(child)
  def die(self):
    self.is_alive = False

```

```python
class Monarchy:
  def __init__(self,king_name):
    self.root = T_node(king_name)

  def birth(self,child_name, parent_name ):
      node =  self.dfs_find(self.root,parent_name)
      if node:
        node.new_birth(child_name)
        print(node.name ,  "  has children ", [ch.name for ch in  node.children])
      else:
        print("could not find the father")

  def death(self,name):
    node =  self.dfs_find(self.root,name)
    if node:
      node.die()
    else:
      print("no one called ",name)
  def get_order_of_succecors(self):
      out = []
      self.dfs(self.root,out)
      return out
  def dfs(self,node,out):
    if node.is_alive:
      out.append(node.name)
    for child in node.children:
      self.dfs(child,out)
  def dfs_find(self,node,name):
    print(node.name , name, node.name==name)
    if node.name == name:
      return node
    else:
      for child in node.children:
        print(child.name , "from for dfs")
        return self.dfs_find(child,name)


```
