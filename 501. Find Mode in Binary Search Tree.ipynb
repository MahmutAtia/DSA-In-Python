```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def findMode(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        all_nodes ={}
        # edge case
        if not root:
            return [0]

        self.dfs(root,all_nodes)
        maks = max(list(all_nodes.values()))

        out = [key for key,val in all_nodes.items() if val == maks ]
        return out
    def dfs(self,node,all_nodes):


        if not all_nodes.get(node.val):
            all_nodes[node.val] = 1
        else:
            all_nodes[node.val]+=1


        if node.left:
            self.dfs(node.left,all_nodes)

        if node.right:
            self.dfs(node.right,all_nodes)

```
