```python
class Solution:
    def isValid(self, s: str) -> bool:
        li = []
        for i,ch in enumerate(s):

            if ch == "(":
                li.append(")")

            elif ch == "[":
                li.append("]")
            elif ch == "{"  :
                li.append("}")
            else:
                if li:
                    if li.pop() == ch:
                        pass

                    else:
                        return False
                else:
                    return False

        if li:
            return False
        else:
            return True
```
