```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        i = len(cost)
        min_costs = [False for _ in cost]
        return min(self.min_cost(cost,i-1,min_costs),self.min_cost(cost,i-2,min_costs))

    def min_cost(self,cost,i,min_costs):


        # print(cost[i],i)
        if i<0:
            return 0
        if i == 0 or i == 1:
            return  cost[i]
        if not min_costs[i-1]:
            min_costs[i-1] = self.min_cost(cost,i-1,min_costs)
        if not min_costs[i-2]:
            min_costs[i-2]= self.min_cost(cost,i-2,min_costs)
        return cost[i] +  min(min_costs[i-1],min_costs[i-2])
```
