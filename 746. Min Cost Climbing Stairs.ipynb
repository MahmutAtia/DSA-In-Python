```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        i = len(cost)
        return min(self.min_cost(cost,i-1),self.min_cost(cost,i-2))

    def min_cost(self,cost,i):


        # print(cost[i],i)
        if i<0:
            return 0
        if i == 0 or i == 1:
            return  cost[i]
        return cost[i] +  min(self.min_cost(cost,i-1),self.min_cost(cost,i-2))

```
