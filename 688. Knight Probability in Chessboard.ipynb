```python
class Solution:
    def knightProbability(self, n: int, k: int, row: int, column: int) -> float:
        directs =[
            #up
            [-2,1],
            [-2,-1],
            #down
            [2,1],
            [2,-1],
            #left
            [-1,-2],
            [1,-2],
            #right
            [1,2],
            [-1,2]

        ]
        db = {}
        all = []
        if k==0:
            return 1
        k=2
        self.rec(n,k,row,column,db,all,directs)
        print(len(all))
        print(db)
        return sum(all)/len(all)


    def rec(self,n,k,row,col,db,all,directs):
        k -= 1 # making a move
        li = []



        for dir in directs:
            r = row + dir[0]
            c= col + dir[1]

            if r <0 or c < 0 or r >= n or c >= n:
                li.append(False)
            else:
                li.append(True)


            if k > 0: #still have moves
                if not db.get((r,c)) :
                    li =  self.rec(n,k,r,c,db,all,directs)
                    db[(r,c)] = li
                else:
                    li = db[(r,c)]



        all+=li
        return li
        # # after for
        # db[(row,col)] = li


```
