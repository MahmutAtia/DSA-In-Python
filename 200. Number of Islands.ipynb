```python
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        direcs =[
            [-1,0], #up
            [0,1], #right
            [1,0] ,#down
            [0,-1] # left
        ]
        seen = self.get_seen_grid(grid)
        queue = [[0,0]]
        while queue:
            curr = queue.pop(0)

            i = curr[0]
            j= curr[1]


            if i< 0 or j<0 or i>=len(grid) or j >= len(grid[0]) or seen[i][j]:
                print(i,j,seen[i][j], "pass")

                continue


            if grid[i][j] == "1":
                    seen[i][j] = "l"
            else:
                    seen[i][j] = "w"


            for direc in direcs:
               col= i + direc[0]
               row =  j +direc[1]

               queue.append([row,col])
            print(queue)


        self.print_seen(seen)
        return 3
    def get_seen_grid(self,grid):

       #  print(len(grid),len(grid[0]),)
        seen = [[False ] * len(grid[0]) ] * len(grid)
       # print(seen)
        return seen

    def print_seen(self,seen):
        for li in seen:
            print(li)

```
