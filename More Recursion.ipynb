### 1- Revirse String

```python
# Try 1:
def foo(s,count = 0):


  #base
  if len(s)-1 == count:
    print(count)
    return s[count]

  return foo(s,count+1) + s[count]
```

```python
foo("mahmoud")
```

### 3- Decimal To Binery

```python
def d2b(num):
  print(num)
  #base
  if num == 0:
    return ''

  return   d2b(num//2) + str(num%2)

```

### 4- Sum of nutrual numbers


```python
def foo(num):
  print(num)

  # base
  if num == 0:
    return 0

  return num + foo(num-1)
```

```python
foo(10), foo(5)
```

### 4- Fibinachi

```python
def foo(n):
  if n == 1 or n == 0:
    return n

  return foo(n-1) + foo(n-2)
```

```python
foo(10)
```

```python
# optimized try 1
db = {}
def foo(n,db):
    print(db)
    if n == 1 or n == 0:
       return n

    if not db.get(n):
      db[n] = foo(n-1,db) + foo(n-2,db)


    return db[n]

```
