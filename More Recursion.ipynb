### 1- Revirse String

```python
# Try 1:
def foo(s,count = 0):


  #base
  if len(s)-1 == count:
    print(count)
    return s[count]

  return foo(s,count+1) + s[count]
```

### 3- Decimal To Binery

```python
def d2b(num):
  print(num)
  #base
  if num == 0:
    return ''

  return   d2b(num//2) + str(num%2)

```
