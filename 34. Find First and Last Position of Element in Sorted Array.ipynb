```python
    def searchRange(nums, target) :
        l = 0
        r = len(nums)-1


        while l<r:
            mid = (l+r) // 2
            print(l,r)
            print(mid)
            if nums[mid] < target:
                l= mid+1
            elif nums[mid] > target:
                r = mid-1

            elif nums[mid] == target:
                break

            else:
                return [-1,-1]

        # we found the target number
        idx_l = mid
        idx_r = mid

        left,right = True , True
        while left or right :
            print(idx_l,idx_r)
            if nums[idx_l-1] == nums[mid]:
                idx_l-=1
            else:
              left = False

            if nums[idx_r+1] == nums[mid]:
                idx_r += 1

            else:
              right = False



        return [idx_l,idx_r]
```
