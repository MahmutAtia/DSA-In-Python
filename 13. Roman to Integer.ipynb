```python
class Solution:
    def romanToInt(self, s: str) -> int:
        hash_map = {
            "I":1,
            "V":5,
            "IV":4,
            "X":10,
            "IX":9,
            "L":50,
            "XL":40,
            "C":100,
            "XC":90,
            "D":500,
            "CD":400,
            "M":1000,
            "CM":900

        }

        p1 = 0
        p2 = 1
        total = 0
        while p1 < len(s):
            if p2==len(s):
                total += ( hash_map[s[p1]])
                p1+=1

            elif hash_map.get(s[p1]+s[p2]):
                    total+= ( hash_map[s[p1]+s[p2]])
                    p1+=2
                    p2+=2

            else :
                total += ( hash_map[s[p1]])
                p1+=1
                p2+=1
        return total
```
