```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:

        # edge caces
        if head  == None:
            return None
        if head.next == None:
            return head


        prev = None
        current_node = head
        next_node = None
        while current_node:

            next_node = current_node.next # a new head is made the reftence was cut
            current_node.next = prev
            prev = current_node

            current_node = next_node # we move to the new head

        return prev
```
