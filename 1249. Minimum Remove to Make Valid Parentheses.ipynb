```python
class Solution:
    def minRemoveToMakeValid(self, s: str) -> str:
        stack_open = []
        stack_close = []
        for i,ch in enumerate(s):
            if ch == "(":
                stack_open.append(i)
            elif ch ==")":
                if stack_open:
                    stack_open.pop()
                else:
                    stack_close.append(i)



        new_stack = stack_open + stack_close # indexs to remove
        if new_stack:
            li = list(s)
            out = ""
            for i in range(len(li)):
                if i not in new_stack:
                    out += li[i]




            return out
        else:
            return s
```
