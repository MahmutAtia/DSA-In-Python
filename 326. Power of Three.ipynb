```python
class Solution(object):
    def isPowerOfThree(self, n):
        """
        :type n: int
        :rtype: bool
        """
        return self.f(n)
    def f(self,n,three_pow_x = 1):
        # base
        if n == three_pow_x:
            return True
        if n < three_pow_x:
            return False

        return self.f(n,three_pow_x*3)
```
