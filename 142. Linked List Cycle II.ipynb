```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:

    # floyd's Algorithm
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next or not head.next.next : return None #edge cace
        slow_curr = head.next
        fast_curr = head.next.next

        not_found = True
        while not_found:

            if slow_curr == fast_curr:
                slow_curr = head
                not_found = False


            else:
                if slow_curr.next and fast_curr.next:
                    if fast_curr.next.next:
                        slow_curr = slow_curr.next
                        fast_curr = fast_curr.next.next
                    else:
                        return None
                else:
                    return None

        while slow_curr != fast_curr:
            slow_curr = slow_curr.next
            fast_curr = fast_curr.next

        return slow_curr

```
