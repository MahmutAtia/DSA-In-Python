```python
class Solution:


    def orangesRotting(self, grid: List[List[int]]) -> int:
       # edge case 1 , there is no fresh
       if not self.find_fresh(grid):
           return 0

       # edge case 2 there is no rotten
       rotten_li =  self.find_rotten(grid)
       if not rotten_li:
           return -1

       else:
           return self.mins_calc(grid)





    def find_rotten(self,grid):
        li = self.loop_over_grid(grid,2) # 2 is the rotten value
        return li

    def find_fresh(self,grid):
        li = self.loop_over_grid(grid,1) # 1 is the fresh value
        if li:
            return True
        else:
            return False



    def mins_calc(self,grid):
        count = 0
        len_rot = 0
        while self.find_fresh(grid):
            rotten_li = self.find_rotten(grid)
            # if there is no change in rotten list and we still
            if len_rot == len(rotten_li):
                return -1

            self.turn_to_rotten(rotten_li,grid)
            count +=1
            len_rot = len(rotten_li)
        return count



    def turn_to_rotten(self,rotten_li,grid):
        for i in rotten_li:
            self.bfs(i,grid)

    def bfs(self, grid_idx,grid):
        i = grid_idx[0]
        j = grid_idx[1]

        # directions
        direcs= [
            [-1,0], #up
            [0,1], # right
            [1,0], #down
            [0,-1]
        ]

        for direc in direcs:
            row = i+direc[0]
            col = j+direc[1]

            if row < 0 or row >= len(grid) or col < 0 or col>=len(grid[0]) or grid[row][col] == 0 or grid[row][col] == 2:
                continue
            else:
                grid[row][col] = 2

    def loop_over_grid(self, grid,value):
        li = []
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == value:
                    li.append([i,j])
        return li

```

```python
class Solution:


    def orangesRotting(self, grid: List[List[int]]) -> int:
        li_rotten, fresh = self.scan_grid(grid)

        #edge 1- no fresh
        if not fresh:
            return 0

        # edge 2- no fresh
        if not li_rotten:
            return -1

        # print(li_rotten, fresh)
        # normal circumstance
        fresh , mins = self.bfs(grid,li_rotten,fresh)

        if fresh:
            return -1
        else:
            return mins


    def bfs(self, grid,li_rotten,fresh):  #fresh is the number of fresh
        queue = li_rotten  # initial queue
        len_queue = len(queue)
        mins = 0 # minutes logic



        # directions
        direcs= [
            [-1,0], #up
            [0,1], # right
            [1,0], #down
            [0,-1]
        ]
        while queue:
            print("queue ", queue , "fresh :", fresh, "mins:",mins,"len_q :", len_queue)
            curr = queue.pop(0)
            len_queue-=1 #decriment

            for direc in direcs:
                row = curr[0]+direc[0]
                col = curr[1]+direc[1]

                if row < 0 or row >= len(grid) or col < 0 or col>=len(grid[0]) or grid[row][col] == 0 or grid[row][col] == 2:
                    # print("inif", row,col)
                    continue
                else: # must be equal to 1
                    # print("inelse", row,col)

                    grid[row][col] = 2
                    queue.append([row,col])
                    fresh-=1

            # dealing with level and minuts logic
            if len_queue == 0:
                mins+=1
                len_queue= len(queue)
        #print(fresh , mins)
        return fresh , mins-1 # mins because the last 1 in the queue is already rotten


    def scan_grid(self, grid):
        li_rotten = []
        fresh = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 2:
                    li_rotten.append([i,j])
                elif grid[i][j] == 1:
                    fresh +=1

        return li_rotten , fresh
```
