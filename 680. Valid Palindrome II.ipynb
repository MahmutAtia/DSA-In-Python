```python
class Solution:
    def validPalindrome(self, s: str) -> bool:
        s = s.lower()
        s_final = re.sub("\W+","",s).replace("_","")



        # initalize pointers
        pl = 0
        pr = len(s_final)-1
        out = True
        while pr-pl-1 >=0:


            if s_final[pl] == s_final[pr]:

                out = True

            else:

                return self.valid(s_final,pl+1,pr) or self.valid(s_final,pl,pr-1)

            # moving pointerrs
            pl+=1
            pr -=1

        return out

    def valid(self,s_final , pl,pr):

        out = True
        while pr-pl-1 >=0:
            if s_final[pl] == s_final[pr]:
                out = True
            else:
                return False
            pl+=1
            pr -=1
        return out

```
